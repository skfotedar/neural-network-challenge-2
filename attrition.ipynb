## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from sklearn.preprocessing import OneHotEncoder
import tensorflow as tf

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique().sort_values(ascending=False)
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
X_df = attrition_df[['HourlyRate', 'Age', 'PerformanceRating',
                               'YearsAtCompany','DistanceFromHome',
                                'EnvironmentSatisfaction',
                               'YearsWithCurrManager','YearsSinceLastPromotion',
                               'PercentSalaryHike','MaritalStatus' ]]
X_df.head()


# Create X_df using your selected columns


# Show the data types for X_df
print(X_df.dtypes)

```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
ohe = OneHotEncoder(handle_unknown='ignore', sparse=False).set_output(transform="pandas")
ohetransform = ohe.fit_transform(X_train[['MaritalStatus','PerformanceRating','EnvironmentSatisfaction']])
ohetransform.head()
X_train = pd.concat([X_train, ohetransform], axis=1).drop(columns=['MaritalStatus','PerformanceRating','EnvironmentSatisfaction'])
X_train.head()

ohetransform = ohe.fit_transform(X_test[['MaritalStatus','PerformanceRating','EnvironmentSatisfaction']])
ohetransform.head()
X_test = pd.concat([X_test, ohetransform], axis=1).drop(columns=['MaritalStatus','PerformanceRating','EnvironmentSatisfaction'])
X_test.head()
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_scaler = scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

```

```python
# Create a OneHotEncoder for the Department column
ohe = OneHotEncoder(handle_unknown='ignore', sparse=False).set_output(transform="pandas")

# Fit the encoder to the training data
ohetransform = ohe.fit_transform(y_train[['Department']])
ohetransform.head()

# Create two new variables by applying the encoder
# to the training and testing data
y_train = pd.concat([y_train, ohetransform], axis=1).drop(columns=['Department'])
y_train
```

```python
# Create a OneHotEncoder for the Attrition column
ohe = OneHotEncoder(handle_unknown='ignore', sparse=False).set_output(transform="pandas")

# Fit the encoder to the training data
ohetransform = ohe.fit_transform(y_train[['Attrition']])
ohetransform.head()

# Create two new variables by applying the encoder
# to the training and testing data
y_train = pd.concat([y_train, ohetransform], axis=1).drop(columns=['Attrition'])
y_train.head()
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
columns = X_train_scaled.shape[1]

# Create the input layer
input_layer = Input(shape=(columns,), name='input_features')

# Create at least two shared layers
shared_layer1 = Dense(64, activation='relu')(input_layer)
shared_layer2 = Dense(32, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer


# Create the output layer
department_output = Dense(3, activation='softmax', name='department_output')(shared_layer2)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer


# Create the output layer
attrition_output = Dense(3, activation='softmax', name='attrition_output')(shared_layer2)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[department_output, attrition_output])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(
    X_train,y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2
)
```
